```python
import pandas as pd

# Cargar el archivo CSV
ruta = '../data/WA_Fn-UseC_-Telco-Customer-Churn.csv'
df = pd.read_csv(ruta)

df.head()
```

```python
df.info()
```

```python
df['TotalCharges'].head(10)
```

```python
df[df['TotalCharges'].str.strip() == '']
```

```python
df = df[df['TotalCharges'].str.strip() != '']
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'])
```

```python
df['TotalCharges'].dtype
```

```python
df['Churn'].value_counts()
```

```python
import matplotlib.pyplot as plt

# Contar valores de Churn
churn_counts = df['Churn'].value_counts()

# Gráfico de barras
churn_counts.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Distribución de Churn')
plt.ylabel('Cantidad de clientes')
plt.show()
```

```python
# Tabla cruzada
pd.crosstab(df['Contract'], df['Churn'])

# Gráfico de barras apiladas
pd.crosstab(df['Contract'], df['Churn']).plot(kind='bar', stacked=True, color=['skyblue','salmon'])
plt.title('Churn según tipo de contrato')
plt.ylabel('Cantidad de clientes')
plt.show()
```

```python
# Separar clientes que se fueron y los que se quedaron
churn_yes = df[df['Churn'] == 'Yes']
churn_no = df[df['Churn'] == 'No']

# Histograma
plt.hist([churn_no['tenure'], churn_yes['tenure']], bins=30, color=['skyblue','salmon'], label=['No','Yes'])
plt.title('Distribución de tenure según Churn')
plt.xlabel('Meses con la compañía')
plt.ylabel('Cantidad de clientes')
plt.legend()
plt.show()
```
